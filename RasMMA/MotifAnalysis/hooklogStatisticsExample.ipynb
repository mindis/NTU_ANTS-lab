```python
# input 
in_tag = "morstar_f"
in_parseFirstPar = True
in_window = 1
in_upgma_threshold = 0.4

# auto-input
in_hl_list_pkf = 'pickle/hl_list_'+ in_tag + '.pickle'
in_group_dict_pkf = 'pickle/group_dict_' + in_tag + '.pickle'

import pickle
with open(in_hl_list_pkf, 'rb') as f:
    in_hl_list = pickle.load(f)    

with open(in_group_dict_pkf, 'rb') as f:
    in_group_dict = pickle.load(f)

# output
# outfile
```

```python
# Example
hash_set = set()
for hl in in_hl_list:
    hashname = hl.split("_")[0]
    hash_set.add(hashname)

print(len(hash_set), "malwares")
print(len(in_hl_list), "processes")

api_set = set()
for g in in_group_dict:
    group_index = g
    height = in_group_dict[group_index][0]
    member_set = in_group_dict[group_index][1]

    print("==")
    print("G-%d, size %d" % (g, len(member_set)))

    group_api_set = set()
    group_common_api_set = set()
    for m in member_set:
        pathfilename = in_hl_list[m]
        hk3 = Hooklog(pathfilename, in_parseFirstPar)
        hk3_set = hk3.getWinSet(in_window)
        api_set.update(hk3_set)
        group_api_set.update(hk3_set)
        group_common_api_set = group_common_api_set.intersection(hk3_set) if len(group_common_api_set) != 0 else hk3_set
        print("  ", m, hk3.digitname)
        print("     len: %d; api_set: %d" % (hk3.length, len(hk3_set)))
        
    print("G-%d, distinct API %d, common API %d." % (g, len(group_api_set), len(group_common_api_set)))
    
    # use last hk3_set to find distinct api
    for api in hk3_set - group_common_api_set:
        print("     api", api)

print("==")
print("Total %d distinct APIs." % (len(api_set)))
```
